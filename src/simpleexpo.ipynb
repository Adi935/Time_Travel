```python
#import default dataset for univariate series
from sktime.datasets import load_airline
```

```python
import pandas as pd
import numpy as np
```

```python
y = load_airline()
print(type(y))
```

```python
y.plot.line()
```

```python
#importing simple exponential smoothing from sktime
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
```

```python
# By default calls model of simple exponential smoothing
forecaster = ExponentialSmoothing()
```

```python
#specifying the forecast horizon
fh = np.arange(1,3)
```

```python
#Fitting the data
forecaster.fit(y=y,fh=fh)
```

```python
# Storing the predicted values in y_pred

y_pred = forecaster.predict()
```

```python
# Will give flat forecast
y_pred.plot.line()
```
